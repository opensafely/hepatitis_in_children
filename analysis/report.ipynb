## Increase in hepatitis cases in children: a rapid OpenSAFELY analysis to support the national incident response

Since January 2022, there have been 74 reported cases of hepatitis in children in the UK in which the common viruses that cause hepatitis have not been detected, but links to SARS-CoV-2 infection are being investigated. UKHSA have initiated a [national incident](https://www.gov.uk/government/news/increase-in-hepatitis-liver-inflammation-cases-in-children-under-investigation) and alerted the public to be on the lookout for signs and symptoms in children.

The aim of this report is to describe the use OpenSAFELY to provide contextual information on children presenting in general practice for symptoms such as gastrointestinal illness and tests that GPs may order (e.g. liver functions) to diagnose hepatitis to inform the response and any investigations.
